```python
def traverse_array(arr):
    """Function to traverse and print all elements in the array."""
    print("Traversing the array:")
    for i, element in enumerate(arr):
        print(f"Index {i}: {element}")

def reverse_array(arr):
    """Function to reverse the array and print the result."""
    print("Reversing the array:")
    reversed_arr = arr[::-1]
    for i, element in enumerate(reversed_arr):
        print(f"Index {i}: {element}")

def sort_array(arr):
    """Function to sort the array and print the result."""
    print("Sorting the array:")
    sorted_arr = sorted(arr)
    for i, element in enumerate(sorted_arr):
        print(f"Index {i}: {element}")

def insert_element(arr, element):
    """Function to insert an element into the array and print the result."""
    print(f"Inserting element {element} into the array:")
    arr.append(element)
    for i, e in enumerate(arr):
        print(f"Index {i}: {e}")

def modify_element(arr, index, new_element):
    """Function to modify an element at a specific index in the array."""
    if 0 <= index < len(arr):
        print(f"Modifying element at index {index} from {arr[index]} to {new_element}:")
        arr[index] = new_element
        for i, e in enumerate(arr):
            print(f"Index {i}: {e}")
    else:
        print("Invalid index")

def main():
    # Initialize an array
    arr = [10, 20, 30, 40, 50]
    print("Original array:", arr)

    while True:
        print("\nMenu:")
        print("1. Traverse array")
        print("2. Reverse array")
        print("3. Sort array")
        print("4. Insert element")
        print("5. Modify element")
        print("6. Exit")
        choice = input("Enter your choice (1-6): ")

        if choice == '1':
            traverse_array(arr)
        elif choice == '2':
            reverse_array(arr)
        elif choice == '3':
            sort_array(arr)
        elif choice == '4':
            element = int(input("Enter the element to insert: "))
            insert_element(arr, element)
            sort_array(arr)
        elif choice == '5':
            index = int(input("Enter the index to modify: "))
            new_element = int(input("Enter the new element: "))
            modify_element(arr, index, new_element)
        elif choice == '6':
            print("Exiting the program.")
            break
        else:
            print("Invalid choice. Please enter a number between 1 and 6.")

if __name__ == "__main__":
    main()

```
